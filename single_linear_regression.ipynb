```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/datasets/spotify_tracks.csv')
```

```python
# Display the first few rows of the dataset to understand its structure
df.head()
```

```python
# Display dataset information to check column types and missing values
df.info()
```

```python
# Select relevant features for linear regression
X = df['valence'].values  # Feature: Valence
y = df['danceability'].values  # Target: Danceability
```

```python
# Calculate and display the correlation coefficient to assess linearity
corr, _ = pearsonr(X, y)
print(f"Correlation: {corr}")
```

```python
# Visualize the correlation matrix for numeric features
numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', annot_kws={"fontsize": 8})
plt.title("Correlation Matrix")
plt.show()
```

```python
# Scatter plot to visualise the relationship between Valence and Danceability
plt.figure(figsize=(10, 8))
plt.scatter(X, y, alpha=0.7)
plt.xlabel("Valence")
plt.ylabel("Danceability")
plt.title('Scatter Plot: Valence vs Danceability')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
```

```python
# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Visualize the train-test split
plt.figure(figsize=(10, 8))
plt.scatter(X_train, y_train, label='Training Data', color='r', alpha=0.5)
plt.scatter(X_test, y_test, label='Testing Data', color='g', alpha=0.5)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.legend()
plt.title("Train-Test Split Visualization")
plt.show()
```

```python
# Reshape training and testing data for compatibility with sklearn
X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
```

```python
# Create and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict danceability for the test set
y_pred = model.predict(X_test)
```

```python
# Visualize model predictions against actual test data
plt.figure(figsize=(10, 8))
plt.plot(X_test, y_pred, label='Predicted Line', color='r')
plt.scatter(X_test, y_test, label='Actual Test Data', color='g', alpha=0.7)
plt.xlabel("Valence")
plt.ylabel("Danceability")
plt.title('Model Predictions vs Actual Test Data')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.legend()
plt.show()
```

```python
# Predict Danceability of music with 0.821 valence
model.predict(np.array([[0.821]]))[0]
```

```python
# Evaluate the model using MSE and RMSE
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"\nMean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
